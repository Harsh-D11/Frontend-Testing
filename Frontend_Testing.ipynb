Q. Use tools like JMeter or Gatling to perform load testing on a web
application and analyze the results to identify performance bottlenecks.?

Configured a JMeter Thread Group with 5 virtual users, ramp‑up 5 seconds, 1 loop, targeting https://www.google.com/ and https://blazedemo.com/.​​

Added a Summary Report listener and executed the test; the report showed 15 total requests with non‑zero throughput but 100% errors, indicating that the public sites rejected automated requests, which is a real‑world bottleneck caused by external rate limiting, not your client machine.​

- www.google.com: 10 samples, Error 100%, Throughput about 1.0/min.
- www.blazedemo.com: 5 samples, Error 100%, Throughput about 1.2/sec.
- TOTAL: 15 samples, overall throughput about 1.4/min.
